This file is used to generate the boxplots found in the final report.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning) 
```

```python
df_joined = pd.read_csv('spreadsheets/joined.csv')
df_radiomics = pd.read_csv('spreadsheets/Radiomics_Reduced.csv')
df_annotations = pd.read_csv("spreadsheets/BUSL_Notes.csv", skiprows=1)
df_annotations = df_annotations.dropna(axis="rows", how="all").dropna(axis="columns", how="all")

# Add in a column for filename, to match up the proper images
df_annotations["filename"] = df_annotations['Sample name'].str.split(expand=True)[0].str[2:]
```

Abruptness and Average Max Gradient


```python
excluded_vals = ['Not circumscribed']
df_joined_filtered = df_joined[~df_joined.isin(excluded_vals)]
axes = df_joined_filtered.boxplot(column=['abruptness'], by=['Marginal ZoneACR'], grid=False, figsize=(12, 4))
fig = axes.get_figure()
fig.suptitle('')
axes = df_joined_filtered.boxplot(column=['avg_max_gradient'], by=['Marginal ZoneACR'], grid=False, figsize=(12, 4))
fig = axes.get_figure()
fig.suptitle('')
```
